```python
dates = pd.date_range(start = '2023-01-01', periods =5 , freq = 'D')
print(dates)
```

```python
eu = [10,12,11,15,14]
```

```python
eseries = pd.Series(data = eu, index = dates, name = "Daily_use")
```

```python
print(eseries)
```

```python
data_reg = {
    'R1':[10, 12 ,14,15, 11],
    'R2':[20, 22 , 25, 24,20]
}

e_df = pd.DataFrame(data = data_reg, index = dates)
print(e_df)
```

```python
print("Information about our DataFrame:")
e_df.info() # Gives data types, non-null counts
print("\n")

print("Descriptive statistics:")
print(e_df.describe()) # Gives mean, std, min, max, etc.
print("\n")

print("First few rows (head):")
print(e_df.head(3)) # Show first 3 rows
print("\n")

print("Accessing a specific column (e.g., RegionA_Load):")
print(e_df['R1'])
```

```python
import matplotlib.pyplot as plt

e_df.plot(figsize = (10,5))

plt.ylabel("Load (units)")
plt.xlabel("Date")
plt.grid(True)
plt.show()
```
